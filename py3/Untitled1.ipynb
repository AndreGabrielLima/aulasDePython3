```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent',
            "McDonald's", 'Visa',
            'Facebook', 'Ailababa',
            'Louis Vuitton']
bilhoes = [947062, 819573, 705646,
           611460, 214023, 196526,
           191032, 186421, 169966,
           124273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
plt.rcParams['figure.figsize'] = (15,8)

plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show()
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color = 'green', lw = 10)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
plt.plot(empresas, bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilhões', fontsize = 20)
plt.legend(['Bilhões de dólares'], loc = 'upper right', fontsize=15)
plt.grid(True)
plt.show()
```
