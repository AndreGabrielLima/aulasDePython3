```python
url_base = 'https://lista.mercadolivre.com.br/'
```

```python
from numpy.lib.function_base import append
produto_nome = input('Qual produto você procura? ')
response = requests.get(url_base + produto_nome)
response.text
site= BeautifulSoup(response.text, 'html.parser')
produtos = site.findAll('div', attrs = {'class': 'ui-search-result__wrapper'})


a = []
for produto in produtos: 
    links = produto.find('a', attrs= {'class','ui-search-link'})
    real = produto.find('span', attrs= {'class': 'price-tag ui-search-price__part shops__price-part'})
    preco = real.text.split()[0]
    print('Link do produto: ', links['href'])
    print('Preço do produto: R$',preco)
    valor = preco
    a+=[valor]
    print('\n')
```

```python
dados = pd.DataFrame(a)
dados.columns = ['Preço']
```

```python
dados
```

```python
dados['Preço'] = pd.to_numeric(dados['Preço'])
dados.head()
```
