```python
url_base = "https://www.americanas.com.br/busca/"
```

```python
from numpy.lib.function_base import append
produto_nome = input('Qual o produto que você procura? ')
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.0.0 Safari/537.36'}
response = requests.get(url_base + produto_nome, headers = headers)
response.text
site = BeautifulSoup(response.text, 'html.parser')
produtos = site.findAll('div', attrs = {'class': "src__Wrapper-sc-1l8mow4-0 fsViFX"})

a = []
for produto in produtos:
    link = produto.find('a', attrs = {'class', 'inStockCard__Link-sc-1ngt5zo-1 JOEpk'})
    real = produto.find('span', attrs = {'class': "src__Text-sc-154pg0p-0 price__PromotionalPrice-sc-h6xgft-1 ctBJlj price-info__ListPriceWithMargin-sc-1xm1xzb-2 liXDNM"})
    preco = real.text.split()[1]
    url_inicio = 'https://www.americanas.com.br'
    url_completo = url_inicio + link['href']
    print('Link do produto: ', url_completo)
    print('Preço do produto: R$', preco) 
    valor = preco
    a +=[valor]
    print('\n')
```

```python
dados = pd.DataFrame(a)
```

```python
dados.columns = ['Preço']
```
